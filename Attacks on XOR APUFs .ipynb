# Modeling Attack on 16-XOR APUF

```python
import numpy
import pandas as pd
import keras_metrics
import keras

from keras.models import Sequential
from keras.layers import Dense,Dropout

from sklearn import model_selection
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import confusion_matrix

from keras.callbacks import TensorBoard 
from keras.callbacks import EarlyStopping


df1=pd.read_csv('./o_16xor.csv',header=None)                 // The feature vectors, the transformation form of challenge C
df2=pd.read_csv('./r_16xor_onehot.csv',header=None)         //  The labels of The synthetic feature corresponding to challenge C

tbCallBack = keras.callbacks.TensorBoard(log_dir='./Graph', 
histogram_freq=1, 
write_graph=True, 
write_images=True)
numpy.random.seed(42)

X = df1.iloc[:600000,:129]
Y = df2.iloc[:600000,:]




train_features, test_features, train_labels, test_labels = model_selection.train_test_split(X, Y, test_size = 0.2, random_state = 42)
# create model
model = Sequential()

model.add(Dense(102, input_dim=129, activation='relu'))
model.add(Dense(85, activation='relu'))
model.add(Dense(68, activation='relu'))
model.add(Dense(34, activation='softmax'))


model.summary()

# Compile model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy',keras_metrics.precision(), keras_metrics.recall()])
# Fit the model

import time
start = time.clock()

results = model.fit(
 train_features, train_labels,
 epochs=300,
 batch_size = 1000,
 validation_data = (test_features, test_labels),
 callbacks = [EarlyStopping(patience=40,mode='min',verbose=1), 
 tbCallBack]
)
print(time.clock() - start)
# evaluate the model
scores = model.evaluate(test_features, test_labels)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```
